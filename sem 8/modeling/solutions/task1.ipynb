```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Параметры модели
mu = 0.5      # Коэффициент роста
K = 1.0       # Константа насыщения
eps = 0.1     # Коэффициент смертности
S = 1.0       # Концентрация ресурса питания

# Начальное значение и диапазон времени
M0 = 0.1  # Начальная биомасса
t_span = (0, 50)  # Время от 0 до 50
t_eval = np.linspace(t_span[0], t_span[1], 500)  # Точки для расчета

# Уравнение
def biomass_dynamics(t, M):
    return mu * (M * S) / (K + S) - eps * M

# Решение уравнения
solution = solve_ivp(biomass_dynamics, t_span, [M0], t_eval=t_eval)

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(solution.t, solution.y[0], label="M(t)", color="blue")
plt.title("Динамика биомассы M(t)", fontsize=14)
plt.xlabel("Время t", fontsize=12)
plt.ylabel("Биомасса M", fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()
```
